```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
os.chdir('/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run/repeat_placements_april22')
```

```python
df = pd.read_csv('iris_rdkit_place.csv', index_col=0)
# make name column from name removing the number
df['inchi_key'] = df['name'].apply(lambda x: x.rsplit('-', 1)[0])
df
```

```python
# show distribution of mean values of ddG for each name so we see what values we are covering
# group by smiles and get mean of ddG
df_group = df[['inchi_key', '∆∆G', 'comRMSD', 'runtime']]
df_grouped = df_group.groupby('inchi_key').mean()
df_grouped
```

```python
# plot distribution of mean values of ddG
# Set the aesthetic style of the plots
sns.set_theme()
sns.set_style('whitegrid')
# show count at top of each bin
ax = sns.histplot(df_grouped['∆∆G'], bins=20)
# Calculate bin heights to annotate each bin
for p in ax.patches:
    height = p.get_height()  # Get the height of each bin
    ax.annotate(f'{int(height)}',  # This converts height to integer, you can modify formatting if needed
                (p.get_x() + p.get_width() / 2., height),  # This places the text at the center of the bin
                ha='center', va='center',  # Horizontally center align text, vertically center
                xytext=(0, 5),  # 5 points vertical offset
                textcoords='offset points')
```

```python
# plot distribution of mean values of comRMSD
sns.histplot(df_grouped['comRMSD'], bins=20)
```

```python
# plot distribution of mean values of runtime
sns.histplot(df_grouped['runtime'], bins=20)
df_grouped['runtime'].describe()
```

```python
# plot violin plots of 5 random inchi_keys
# get 5 random inchi_keys, non repeating
inchi_keys = df_grouped.index
np.random.seed(0)
random_inchi_keys = np.random.choice(inchi_keys, 10, replace=False)
random_inchi_keys
```

```python
# plot violin plots of ddG of 5 random inchi_keys
df_violin = df_group[df_group['inchi_key'].isin(random_inchi_keys)]
# turn x axis labels vertical

# make plot a lot bigger
plt.figure(figsize=(20, 7))
sns.violinplot(x='inchi_key', y='∆∆G', data=df_violin)
plt.xticks(rotation=90)
# add more y lines

# Compute standard deviation for each group
group_sd = df_violin.groupby('inchi_key')['∆∆G'].std()
# change order ot match violin plot
group_sd = group_sd.reindex(df_violin['inchi_key'].unique())

# Annotate each violin with its standard deviation
for i, (key, sd) in enumerate(group_sd.items()):
    # make sure plotting on top of correct violin
    
    plt.text(i, df_violin['∆∆G'].max(), f'SD: {sd:.2f}', horizontalalignment='center')

# Show the plot
```

```python
# histogram distribution of standard deviations of ddG for each inchi_key
ax = sns.histplot(df_group.groupby('inchi_key')['∆∆G'].std(), bins=20)
plt.xlabel('Standard Deviation of ∆∆G')
# Calculate bin heights to annotate each bin
for p in ax.patches:
    height = p.get_height()  # Get the height of each bin
    ax.annotate(f'{int(height)}',  # This converts height to integer, you can modify formatting if needed
                (p.get_x() + p.get_width() / 2., height),  # This places the text at the center of the bin
                ha='center', va='center',  # Horizontally center align text, vertically center
                xytext=(0, 5),  # 5 points vertical offset
                textcoords='offset points')
```

```python
sd_ddG = df_group.groupby('inchi_key')['∆∆G'].std()
sd_ddG.describe()
```

```python
# calculate median
df_group.groupby('inchi_key')['∆∆G'].std().median()
```

```python
# plot standard deviation of ddG and average RMSD together
# get average RMSD for each inchi_key
comRMSD = df_group.groupby('inchi_key')['comRMSD'].mean()
# make a dataframe with both
df_sd_comRMSD = pd.DataFrame({'sd_ddG': sd_ddG, 'comRMSD': comRMSD})
df_sd_comRMSD
```

```python
sns.scatterplot(x='sd_ddG', y='comRMSD', data=df_sd_comRMSD)
plt.xlabel('Standard Deviation of ∆∆G')
plt.ylabel('Average RMSD')
```
