```python
import pandas as pd
import os
os.chdir('/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run/base_compounds')
from rdkit import Chem
import ast
```

```python
# Load the data
df = pd.read_csv('syndirella_input_MASTER_FINAL_exact.csv')
df
```

```python
# remove duplicates
df = df.drop_duplicates(subset='name')
# reindex
df = df.reset_index(drop=True)
df
```

```python
# explode reactants column
df['reactants'] = df['reactants'].apply(ast.literal_eval)
df
```

```python
rxn_column_order = ['1_reaction', '1_r1_smiles', '1_r2_smiles', '1_r_previous_product', '1_product_smiles', 
                    '2_reaction', '2_r1_smiles', '2_r2_smiles', '2_r_previous_product', '2_product_smiles',
                    '3_reaction', '3_r1_smiles', '3_r2_smiles', '3_r_previous_product', '3_product_smiles',
                    '4_reaction', '4_r1_smiles', '4_r2_smiles', '4_r_previous_product', '4_product_smiles',]
reactant_columns = ['1_r1_smiles', '1_r2_smiles', '2_r1_smiles', '2_r2_smiles', '3_r1_smiles', '3_r2_smiles', '4_r1_smiles', '4_r2_smiles']
reaction_columns = ['1_reaction', '2_reaction', '3_reaction', '4_reaction']
product_columns = ['1_product_smiles', '2_product_smiles', '3_product_smiles', '4_product_smiles']
```

```python
reactants_expanded = df['reactants'].apply(pd.Series)
reactants_expanded
```

```python
# make reactant columns
reactants = pd.DataFrame()
for i in range(4):
    r1_smiles = reactants_expanded[i].apply(lambda x: x[0] if type(x) == tuple else x)
    r2_smiles = reactants_expanded[i].apply(lambda x: x[1] if type(x) == tuple else None)
    reactants[reactant_columns[i*2]] = r1_smiles
    reactants[reactant_columns[i*2+1]] = r2_smiles
reactants
```

```python
# make reaction columns
reactions_expanded = df['reaction_names'].apply(ast.literal_eval).apply(pd.Series)
reactions = pd.DataFrame()
for i in range(4):
    reactions[reaction_columns[i]] = reactions_expanded[i]
reactions
```

```python
reactants_expanded = df['reactants'].apply(pd.Series)
reactants_expanded = reactants_expanded.fillna(value=0)
# make product columns
def check_for_previous_product(x, i):
    if x == 0:
        return None
    if i == 0:
        return None
    if len(x) == 1:
        return None
    return 1
reactants_expanded
```

```python
products = pd.DataFrame()
for i in range(4):
    products[f'{i+1}_r_previous_product'] = reactants_expanded[i].apply(lambda x: check_for_previous_product(x, i))
products
# fill all values that are not 1 with None
products = products.where(products == 1)
products
```

```python
# make product columns
which_product = pd.DataFrame()
def get_product(x, i):
    # you're looking at the reactant column for the next reaction
    if type(x) == str:
        return x
    if type(x) == tuple:
        return x[0]
    return None
for i in range(4):
    if i == 3:
        which_product[f'{i+1}_product_smiles'] = df['smiles']
        break
    which_product[f'{i+1}_product_smiles'] = reactants_expanded[i+1].apply(lambda x: get_product(x, i))
which_product
```

```python
# now concatenate all the dataframes
final_df = pd.concat([df, reactants, reactions, products, which_product], axis=1)
# order columns correctly
final_df = final_df.reindex(columns=rxn_column_order)
final_df
```

```python
# move values from the 4_product_smiles to first x_product_smiles that is None
def fill_in_product(x):
    for i in range(4):
        if pd.isnull(x[f'{i+1}_product_smiles']):
            x[f'{i+1}_product_smiles'] = x['4_product_smiles']
            x['4_product_smiles'] = None
            break
    return x
final_df = final_df.apply(fill_in_product, axis=1)
final_df
```

```python
# add all other columns to the end
final_df = pd.concat([final_df, df.drop(columns=['reactants', 'reaction_names'])], axis=1)
final_df.rename(columns={'smiles': 'base_compound_smiles'}, inplace=True)
final_df.drop(columns=['outputs'], inplace=True)
final_df
```
