# How many batches?
23 open slots on CPU
69/23 = 3 = 3 in each batch

If it takes on average 10 sec per placement, and a week contains 604,800 seconds, then can place 60480 in a week. Therefore, if only placing max 10K per base, then will place 30K per batch which should take around 300,000 seconds or 3.5 days. Therefore, should be able to place all in a week.

```python
df = pd.read_csv('bases/MASTER.csv', index_col=0)
df
```

```python
# create batch column with 3 in each batch
# Create the batch column
df['batch'] = np.tile(np.arange(23), 3)[:69]
df
```

```python
# make sure there are no duplicate names
assert len(df.name) == len(df.name.unique())
```

```python
# create seperate csvs for each batch
for i in range(23):
    df[df.batch == i].to_csv(f'batches/batch_{i}.csv', index=False)
```
