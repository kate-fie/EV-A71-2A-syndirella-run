```python
# generate inchikey for each rs
rs['inchikey'] = rs[0].apply(lambda x: Chem.MolToInchiKey(Chem.MolFromSmiles(x)))
rs
```

```python
def find_reactant(row):
    for i in range(1, 7):
        row_name = 'r' + str(i)
        row_name_result = row_name + '_found'
        if row[row_name] == 0:
            row[row_name_result] = None
        else:
            smiles = ast.literal_eval(row[row_name])
            # get inchikey from smiles
            try:
                print(smiles)
                inchi = Chem.MolToInchiKey(Chem.MolFromSmiles(smiles))
            except:
                inchi = None
            # check if inchikey is in rs
            if inchi in rs['inchikey'].values:
                row[row_name_result] = True
            else:
                row[row_name_result] = False
    return row
```

```python
df = df.apply(find_reactant, axis=1)
df
```
