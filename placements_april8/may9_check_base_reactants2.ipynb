```python
df = pd.read_csv('bases/MASTER.csv', index_col=0)
df
```

```python
rs_data = pd.read_fwf('may9_base_reactants.txt', header=None)
rs_data.rename(columns={0: 'smiles'}, inplace=True)
rs_data['inchikey'] = rs_data.apply(lambda x: Chem.MolToInchiKey(Chem.MolFromSmiles(x['smiles'])), axis=1)
rs_data
```

```python
# Create a dictionary for fast lookup
rs_inchikeys = {row['smiles']: row['inchikey'] for index, row in rs_data.iterrows()}

# Function to convert SMILES to InChIKey
def smiles_to_inchikey(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        return Chem.MolToInchiKey(mol)
    except:
        return None

# Function to explode tuples in the 'reactants' column and add '_found' columns
def explode_and_find(df, reactants_col, rs_inchikeys):
    def parse_reactants(reactants):
        # Convert the string representation of tuples into a list of tuples
        try:
            return ast.literal_eval(reactants)
        except:
            return []

    def is_reactant_found(reactant):
        # if tuple, check both
        if isinstance(reactant, tuple):
            return any(is_reactant_found(r) for r in reactant)
        smiles = reactant if isinstance(reactant, str) else None
        if smiles is None:
            return None
        inchikey = smiles_to_inchikey(smiles)
        return inchikey in rs_inchikeys.values()

    # Apply parsing function
    df[reactants_col] = df[reactants_col].apply(parse_reactants)

    # Determine the maximum number of reactants
    max_reactants = df[reactants_col].apply(lambda x: len(x)).max()

    # Create column names dynamically
    reactant_columns = [f'reactant_{i + 1}' for i in range(max_reactants)]
    found_columns = [f'{col}_found' for col in reactant_columns]

    # Expand tuples into separate columns
    reactants_expanded = df[reactants_col].apply(pd.Series)
    reactants_expanded.columns = reactant_columns

    # Check if reactants are found and create '_found' columns
    found_expanded = reactants_expanded.applymap(is_reactant_found)
    found_expanded.columns = found_columns

    # Concatenate the original DataFrame with the expanded reactants and '_found' columns
    df_expanded = pd.concat([df, reactants_expanded, found_expanded], axis=1)

    return df_expanded, reactants_expanded, found_expanded

# Use the function to expand the 'reactants' column and find presence
df_expanded, reactants_expanded, found_expanded = explode_and_find(df, 'reactants', rs_inchikeys)
```
