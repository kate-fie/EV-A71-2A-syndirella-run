```python
# drop everything except certain columns
to_place = pd.read_csv('bases/iris_placements_with_routes.csv', index_col=0)
keep = ['smiles', 'name', 'frag_small', 'reactants', 'reaction_names', 'num_steps', 'compound_set', 'hits'] 
to_place = to_place[keep]
to_place
```

```python
# drop duplicates not considering compound_set
to_place.drop_duplicates(subset=['smiles', 'name', 'frag_small', 'reactants', 'reaction_names', 'num_steps', 'hits'], inplace=True)
to_place.reset_index(drop=True, inplace=True)
to_place
```

```python
# check duplicates names to make sure they are the same◊
to_place[to_place.duplicated(subset='name', keep=False)]
```

```python
# row 69 should have a different end smiles
to_place.loc[69, 'smiles'] = 'c1cc(NC(=O)CO)c(NC(=O)CC#N)cc1C'
# regenerate name
to_place['name'] = to_place['smiles'].apply(lambda x: rdinchi.InchiToInchiKey(rdinchi.MolToInchi(Chem.MolFromSmiles(x))[0]))
# rows 5 and 6 are the same
to_place.drop(6, inplace=True)
to_place
```

```python
# check there are no duplicates
to_place[to_place.duplicated(subset='name', keep=False)]
```
