```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import rdinchi
```

```python
# get df of base compounds to replace
df = pd.read_pickle('/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run/base_compounds/syndirella_input_MASTER.pkl')
df.drop(columns=['reactants', 'reaction_names', 'num_steps', 'compound_set'], inplace=True)
df
```

```python
df['base'] = [rdinchi.InchiToInchiKey(rdinchi.MolToInchi(Chem.MolFromSmiles(smiles))[0]) for smiles in df['smiles']]
```

```python
df
```

```python
df_errors = pd.read_pickle('/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run/interrogate_failed_bases/bases_df_errors.pkl')
df_errors
```

```python
df_fail = df_errors['minimised'] == False
df_errors_fail = df_errors[df_fail]
df_errors_fail
```

```python
# merge df_errors_fail and df
df_merged = pd.merge(df_errors_fail, df, on='base', how='inner')
df_merged
```

```python
# subset is all columns but 'hits'
df_merged.drop_duplicates(subset=['base', 'smiles', 'minimised', 'batch', 'error', 'base_output_dir'], inplace=True)
df = df_merged
df
```

```python
df['hits'] = df['hits'].apply(lambda x: ' '.join([i + 'A' for i in x]))
df
```
