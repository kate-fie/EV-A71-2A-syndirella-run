```python
# get list of all -base- dirs, then write function that sees if that base dir is found in mols list
all_base_dirs = [os.path.dirname(file) for file in glob.glob("**/*-base-*/", recursive=True)]
```

```python
base_dirs = all_base_dirs[3:]
len(base_dirs)
```

```python
df = pd.DataFrame(columns=['batch', 'base', 'base_output_dir', 'minimised', 'error'])
```

```python
batch = [batch.split('/')[0] for batch in base_dirs]
base = [batch.split('/')[1] for batch in base_dirs]
base_output = [batch.split('/')[-1] for batch in base_dirs]
```

```python
# Check if base directories are found in mols list
def is_base_dir_in_mols(base_dir, mols):
    for mol in mols:
        if base_dir in mol:
            return True
    return False
minimised = [is_base_dir_in_mols(base1, mols) for base1 in base_output]
minimised
```

```python
len(minimised)
```

```python
df['batch'] = batch
df['base'] = base
df['minimised'] = minimised
df['base_output_dir'] = base_output
df
```
