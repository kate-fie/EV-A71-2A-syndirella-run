```python
import pandas as pd
```

```python
df = pd.read_pickle('/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run/base_compounds/wt_designs_MASTER_COMPLETE_FINAL_29Feb.pkl')
df.fillna('NaN', inplace=True)
df
```

```python
# add batch column to df
df['batch'] = df.index // 10
df
```

```python
# make dict of reaction names
reaction_names_dict = {
    'Williamson ether': 'Williamson_ether_synthesis',
    'Suzuki': 'Sp3-sp2_Suzuki_coupling',
    'Nucleophillic substitution': 'Nucleophilic_substitution',
    'Bn-alcohol-deprotection': 'Benzyl_alcohol_deprotection',
    'Si-alcohol deprotection': 'TBS_alcohol_deprotection',
    'Boc deprotection': 'N-Boc_deprotection',
    'Amidation': 'Amidation',
    'TBS_alcohol_deprotection': 'TBS_alcohol_deprotection',
    'Benzyl_alcohol_deprotection': 'Benzyl_alcohol_deprotection'
}
```

```python
# make new syndirella input template
smiles = []
reactants = []
reaction_names = []
num_steps = []
hits = []
compound_set = []
for index, row in df.iterrows():
    num_steps.append(row['no-steps'])
    smiles.append(row[f'reaction-product-smiles-{row["no-steps"]}'])
    reactants_this_one = []
    reaction_names_this_one = []
    for i in range(1, row['no-steps']+1):
        r1 = row[f'reactant-1-{i}']
        r2 = row[f'reactant-2-{i}']
        if r2 == 'NaN':
            r2 = None
        if r1 and r2:
            reactants_this_one.append((r1, r2))
        if r1 and not r2:
            reactants_this_one.append((r1))
        print(row[f'reaction-name-{i}'])
        updated_reaction_name = reaction_names_dict[row[f'reaction-name-{i}']]
        reaction_names_this_one.append(updated_reaction_name)
    assert len(reactants_this_one) == row['no-steps']
    assert len(reaction_names_this_one) == row['no-steps']
    reaction_names.append(reaction_names_this_one)
    reactants.append(reactants_this_one)
    hits.append(row['target-names'].split('-')[0:2])
    compound_set.append(row['target-names'])
```

```python
syndirella_input = pd.DataFrame({'smiles': smiles, 'reactants': reactants, 'reaction_names': reaction_names, 'num_steps': num_steps, 'hits': hits, 'compound_set': compound_set})
syndirella_input.reset_index(drop=True, inplace=True)
syndirella_input
```

```python
syndirella_input.to_pickle('/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run/base_compounds/syndirella_input_MASTER.pkl')
syndirella_input.to_csv('/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run/base_compounds/syndirella_input_MASTER.csv')
```
