```python
df = pd.read_csv('/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run/base_compounds/syndirella_input_MASTER_FINAL_exact.csv')
df
```

```python
# get histogram of number of fragments considered overall
df[['frag1', 'frag2', 'frag3']] = df['hits'].str.split(' ', expand=True)
df
```

```python
metadata = pd.read_csv('/Users/kate_fieseler/PycharmProjects/EV-A71-2A-syndirella-run/fragments/metadata.csv')
metadata_dict = dict(zip(metadata['Long code'], metadata['Code']))
metadata_dict
```

```python
# get list of all fragments
frag1 = df['frag1'].tolist()
frag2 = df['frag2'].tolist()
frag3 = df['frag3'].tolist()
fragments = frag1 + frag2 + frag3
# drop none
fragments = [x for x in fragments if x is not None]
fragments = [metadata_dict[x] for x in fragments]
len(fragments)
```

```python
# get list of fragments only if outputs is 'to_hippo'
hippo_frag1 = df.loc[df['outputs'] == 'to_hippo']['frag1'].tolist()
hippo_frag2 = df.loc[df['outputs'] == 'to_hippo']['frag2'].tolist()
hippo_frag3 = df.loc[df['outputs'] == 'to_hippo']['frag3'].tolist()
hippo_fragments = hippo_frag1 + hippo_frag2 + hippo_frag3
# drop none
hippo_fragments = [x for x in hippo_fragments if x is not None]
hippo_fragments = [metadata_dict[x] for x in hippo_fragments]
len(hippo_fragments)
```

```python
# get list of fragments only if outputs is 'output' or 'extra'
output_frag1 = df.loc[df['outputs'] == 'output']['frag1'].tolist()
output_frag2 = df.loc[df['outputs'] == 'output']['frag2'].tolist()
output_frag3 = df.loc[df['outputs'] == 'output']['frag3'].tolist()
extra_frag1 = df.loc[df['outputs'] == 'extra']['frag1'].tolist()
extra_frag2 = df.loc[df['outputs'] == 'extra']['frag2'].tolist()
extra_frag3 = df.loc[df['outputs'] == 'extra']['frag3'].tolist()
output_fragments = output_frag1 + output_frag2 + output_frag3 + extra_frag1 + extra_frag2 + extra_frag3
# drop none
output_fragments = [x for x in output_fragments if x is not None]
non_outputs_fragments = [metadata_dict[x] for x in output_fragments]
len(non_outputs_fragments)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'fragments' and 'hippo_fragments' are lists of categorical data
fragments_df = pd.DataFrame(fragments, columns=['fragments'])
hippo_fragments_df = pd.DataFrame(hippo_fragments, columns=['fragments'])

# Count the frequency of each category
total_counts = fragments_df['fragments'].value_counts()
hippo_counts = hippo_fragments_df['fragments'].value_counts()

# Make sure every category is represented in both Series (to handle missing categories)
total_counts = total_counts.reindex(index = total_counts.index.union(hippo_counts.index), fill_value=0)

hippo_counts = hippo_counts.reindex(total_counts.index, fill_value=0)

# Calculate the difference to be plotted on top
difference_counts = total_counts - hippo_counts

# Create the bar chart
hippo_counts.plot(kind='bar', color='green', label='Freq. of Frags Successful Base')
difference_counts.plot(kind='bar', color='blue', bottom=hippo_counts, label='Freq. of Frags Unsuccessful Base')

plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'fragments', 'hippo_fragments', and 'non_outputs_fragments' are lists of categorical data
fragments_df = pd.DataFrame(fragments, columns=['fragments'])
hippo_fragments_df = pd.DataFrame(hippo_fragments, columns=['fragments'])
non_outputs_fragments_df = pd.DataFrame(non_outputs_fragments, columns=['fragments'])

# Count the frequency of each category
total_counts = fragments_df['fragments'].value_counts()
hippo_counts = hippo_fragments_df['fragments'].value_counts()
non_outputs_counts = non_outputs_fragments_df['fragments'].value_counts()

# Make sure every category is represented in both Series (to handle missing categories)
all_indices = total_counts.index.union(hippo_counts.index).union(non_outputs_counts.index)
total_counts = total_counts.reindex(all_indices, fill_value=0)
hippo_counts = hippo_counts.reindex(all_indices, fill_value=0)
non_outputs_counts = non_outputs_counts.reindex(all_indices, fill_value=0)

# Calculate the remaining counts (not hippo and not non_outputs)
remaining_counts = total_counts - hippo_counts - non_outputs_counts

# Create the bar chart
hippo_counts.plot(kind='bar', color='green', label='Freq. of Frags To Hippo')
non_outputs_counts.plot(kind='bar', color='orange', bottom=hippo_counts, label='Freg. of Frags No Hippo')
remaining_counts.plot(kind='bar', color='red', bottom=hippo_counts + non_outputs_counts, label='Freq. of Frags Fail')

plt.xticks(rotation=90)
plt.legend()
plt.show()
```
